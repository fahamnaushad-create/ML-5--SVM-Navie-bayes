##1. What is a Support Vector Machine (SVM)?
- Answer:
  
  A Support Vector Machine (SVM) is a supervised learning algorithm used for classification and regression. It finds a boundary (hyperplane) that best separates classes by maximizing the margin ‚Äî the distance between the nearest points of each class and the boundary. SVMs work well for high-dimensional data and can use different kernels to handle non-linear separation. For a beginner: imagine drawing a line between two groups of points such that the line stays as far away as possible from both groups.

##2. What is the difference between Hard Margin and Soft Margin SVM?
- Answer:
  
  Hard-margin SVM requires perfect separation of classes with no training errors ‚Äî it works only if data are perfectly linearly separable. Soft-margin SVM allows some misclassifications (slack) to handle noise and overlapping classes; a penalty parameter balances margin width and errors. In practice, soft-margin is used nearly always because real data are noisy. Think of soft margin as allowing a few misfits to get a sturdier rule that generalizes better.

##3. What is the mathematical intuition behind SVM?
- Answer:
  
  SVM solves an optimization problem to maximize the margin between classes subject to classification constraints; geometrically it looks for the widest possible gap between class-supporting hyperplanes. This becomes a quadratic optimization with linear constraints; the solution depends only on a subset of points (support vectors). The kernel function maps inputs to higher dimensions so linear separation there corresponds to non-linear separation in the original space.

##4. What is the role of Lagrange Multipliers in SVM?
- Answer:
  
  Lagrange multipliers convert the constrained optimization problem (maximize margin under constraints) into an unconstrained dual problem that is easier to solve. In the dual form, classifiers are expressed as weighted sums of training points where weights are the Lagrange multipliers. Only points with nonzero multipliers (support vectors) affect the final model ‚Äî this sparsity is one reason SVMs are efficient at prediction time.

##5. What are Support Vectors in SVM?
- Answer:
  
  Support vectors are the training samples that lie closest to the decision boundary and determine the position of the hyperplane. They have non-zero Lagrange multipliers and directly shape the final model; removing non-support vectors usually does not change the boundary. Intuitively, they are the ‚Äúcritical‚Äù examples the SVM uses to define the separation margin.

##6. What is a Support Vector Classifier (SVC)?
- Answer:
  
  Support Vector Classifier (SVC) is the SVM algorithm applied to classification tasks. It learns a decision boundary (with margin) that separates classes and can be configured for hard or soft margin, and to use kernels for non-linear separation. In scikit-learn, sklearn.svm.SVC implements the classifier version with options for kernel, C, gamma, etc.

##7. What is a Support Vector Regressor (SVR)?
- Answer:
  
  Support Vector Regressor (SVR) adapts SVM ideas to regression: it fits a function within an Œµ-margin tube around the data, penalizing errors outside that tube. The objective balances model flatness and deviations larger than Œµ. Like SVC, SVR can use kernels to model nonlinear relationships while keeping a sparse representation based on support vectors.

##8. What is the Kernel Trick in SVM?
- Answer:
  
  The kernel trick maps data into a higher-dimensional space without computing coordinates explicitly by replacing dot products with a kernel function. This means you can build nonlinear decision boundaries while computing in the original space, using functions like polynomial or RBF kernels. For a beginner: kernels let you ‚Äúbend‚Äù the input space so a straight line in the new space separates the classes.

##9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel?
- Answer:
  
   Linear kernel is just the dot product ‚Äî good when data are roughly linearly separable and fast. Polynomial kernel models polynomial relationships (degree hyperparameter controls complexity) but can overfit for high degrees. RBF (Gaussian) kernel maps points using radial basis functions and handles complex non-linear shapes with relatively few hyperparameters; it's versatile but can be sensitive to gamma. Choose linear for simplicity, RBF for generally flexible non-linearity, polynomial for specific polynomial relationships.

##10. What is the effect of the C parameter in SVM?
- Answer:
  
   C controls the trade-off between a wide margin and classification errors. A small C allows wider margin with more misclassifications (strong regularization), favoring generalization. A large C penalizes misclassification heavily and aims to fit training data closely (risking overfitting). So tune C to find a balance based on validation performance.

##11. What is the role of the Gamma parameter in RBF Kernel SVM?
- Answer:
  
   gamma defines how far the influence of a single training example reaches in RBF kernel: low gamma means wide influence (smoother decision surface), high gamma means narrow influence (more complex, wiggly surface). Too large gamma can overfit, too small can underfit. Use cross-validation to choose a suitable gamma.

##12. What is the Na√Øve Bayes classifier, and why is it called "Na√Øve"?
- Answer:
  
   Na√Øve Bayes is a family of probabilistic classifiers based on Bayes‚Äô theorem with a strong (na√Øve) assumption: all features are conditionally independent given the class. It‚Äôs "na√Øve" because in most real datasets features are not strictly independent, but despite that assumption it often works well in practice, especially for text data. It is simple and computationally efficient.

##14. Explain the differences between Gaussian Na√Øve Bayes, Multinomial Na√Øve Bayes, and Bernoulli Na√Øve Bayes.
- Answer:
  
   Gaussian NB assumes continuous features and models them with a Gaussian distribution per class. Multinomial NB models discrete count data (like word counts in documents). Bernoulli NB models binary features (feature present/absent). Choose the variant based on the data type: continuous ‚Üí Gaussian, counts ‚Üí Multinomial, binary indicators ‚Üí Bernoulli.

##15. When should you use Gaussian Na√Øve Bayes over other variants?
- Answer:
  
   Use Gaussian Na√Øve Bayes when your features are continuous and roughly follow a normal distribution per class (or when that approximation is reasonable). It‚Äôs common with numerical sensor data or physical measurements. If your features are counts or binary, Multinomial or Bernoulli are better suited.

##16. What are the key assumptions made by Na√Øve Bayes?
- Answer:
  
   Main assumptions: (1) conditional independence of features given the class, and (2) that the chosen likelihood model (Gaussian, multinomial, Bernoulli) is appropriate for the feature distribution. Also assumes training examples are independently sampled. These simplify computation but may not hold exactly in real data.

##17. What are the advantages and disadvantages of Na√Øve Bayes?
- Answer:
  
   Advantages: very fast to train and predict, needs few training examples, works well with high-dimensional data (like text), and is simple to interpret. Disadvantages: the conditional independence assumption is often false (can hurt accuracy), and it can perform poorly if feature distributions are misspecified or if features are highly correlated.

##18. Why is Na√Øve Bayes a good choice for text classification?
- Answer:
  
   Text data are high-dimensional (thousands of words) and sparse; Na√Øve Bayes handles this well and trains quickly on counts. The independence assumption is tolerable in practice for bag-of-words models, and Multinomial NB often gives strong baselines for news/topic or spam classification. It also scales well to large corpora.

##19. Compare SVM and Na√Øve Bayes for classification tasks?
- Answer:
  
   SVMs are margin-based and often perform well on small-to-medium datasets and complex boundaries (with kernels), while Na√Øve Bayes is probabilistic, fast, and scales to high-dimensional sparse data like text. SVM is usually more robust in many tabular tasks; Na√Øve Bayes is simpler, faster, and sometimes surprisingly effective for text. Choose based on data size, sparsity, and need for probabilistic outputs.

##20. How does Laplace Smoothing help in Na√Øve Bayes?
- Answer:
  
   Laplace smoothing (add-one smoothing) prevents zero probabilities for unseen feature-class pairs by adding a small constant to counts. This avoids multiplying by zero when computing likelihoods and improves generalization on rare features. For text tasks, this is essential since new words can appear in test data.

##Practical 1. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy.

```python
# Practical 1
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()
X, y = iris.data, iris.target
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

clf = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
clf.fit(Xtr, ytr)
y_pred = clf.predict(Xte)
print("Accuracy:", round(accuracy_score(yte, y_pred), 4))
```

##Practical 4. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary.

```python
# Practical 4
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

iris = datasets.load_iris()
X, y = iris.data[:, :4], iris.target  # full features, reduce later

# For visualization, reduce to 2D with PCA
X2 = PCA(n_components=2, random_state=42).fit_transform(X)
Xtr, Xte, ytr, yte = train_test_split(X2, y, test_size=0.3, random_state=42, stratify=y)

sc = StandardScaler().fit(Xtr)
Xtr_s = sc.transform(Xtr)
Xte_s = sc.transform(Xte)

clf = SVC(kernel='poly', degree=3, C=1.0, gamma='scale').fit(Xtr_s, ytr)

# Plot decision boundary
xx, yy = np.meshgrid(np.linspace(Xtr_s[:,0].min()-1, Xtr_s[:,0].max()+1, 300),
                     np.linspace(Xtr_s[:,1].min()-1, Xtr_s[:,1].max()+1, 300))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.2)
plt.scatter(Xtr_s[:,0], Xtr_s[:,1], c=ytr, edgecolor='k', s=50)
plt.title("Polynomial Kernel SVM (PCA-reduced 2D)")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.show()
```

##Practical 7. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually.

```python
# Practical 7
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

X, y = make_blobs(n_samples=200, centers=2, random_state=6, cluster_std=1.2)
X = X[:, :2]
sc = StandardScaler().fit(X); Xs = sc.transform(X)

Cs = [0.1, 1, 10]
fig, axs = plt.subplots(1, len(Cs), figsize=(15,4))
xx, yy = np.meshgrid(np.linspace(Xs[:,0].min()-1, Xs[:,0].max()+1, 300),
                     np.linspace(Xs[:,1].min()-1, Xs[:,1].max()+1, 300))

for ax, C in zip(axs, Cs):
    clf = SVC(kernel='rbf', C=C, gamma='scale').fit(Xs, y)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.2)
    ax.scatter(Xs[:,0], Xs[:,1], c=y, edgecolor='k')
    ax.set_title(f"C = {C}")
plt.show()
```

##Practical 11. Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C, gamma, kernel).

```python
# Practical 11
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

param_grid = {'C':[0.1,1,10], 'gamma':['scale','auto'], 'kernel':['rbf','linear','poly']}
grid = GridSearchCV(SVC(), param_grid, cv=4, scoring='accuracy', n_jobs=1)
grid.fit(Xtr, ytr)

best = grid.best_estimator_
print("Best params:", grid.best_params_)
print("Test accuracy with best estimator:", round(accuracy_score(yte, best.predict(Xte)),4))
```

##Practical 20. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy.

```python
# Practical 20
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = load_breast_cancer(return_X_y=True)
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

svc = SVC(kernel='linear')  # linear needed for feature ranking
rfe = RFE(svc, n_features_to_select=10).fit(Xtr, ytr)
Xtr_rfe, Xte_rfe = rfe.transform(Xtr), rfe.transform(Xte)

svc_full = SVC(kernel='linear').fit(Xtr, ytr)
svc_rfe = SVC(kernel='linear').fit(Xtr_rfe, ytr)

print("Accuracy full:", round(accuracy_score(yte, svc_full.predict(Xte)),4))
print("Accuracy after RFE:", round(accuracy_score(yte, svc_rfe.predict(Xte_rfe)),4))
```

##Practical 21. Write a Python program to train an SVM Classifier and evaluate its performance using Precision, Recall, and F1-Score instead of accuracy.

```python
# Practical 21
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_fscore_support

X, y = load_breast_cancer(return_X_y=True)
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

clf = SVC().fit(Xtr, ytr)
pred = clf.predict(Xte)
p, r, f, _ = precision_recall_fscore_support(yte, pred, average='binary')
print("Precision:", round(p,4), "Recall:", round(r,4), "F1-score:", round(f,4))
```

##Practical 23. Write a Python program to train an SVM Classifier and visualize the Confusion Matrix using seaborn.

```python
# Practical 23
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

X, y = load_wine(return_X_y=True)
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

clf = SVC().fit(Xtr, ytr)
cm = confusion_matrix(yte, clf.predict(Xte))

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted'); plt.ylabel('True'); plt.title('Confusion Matrix')
plt.show()
```

##Practical 24. Write a Python program to train an SVM Regressor (SVR) and evaluate its performance using Mean Absolute Error (MAE) instead of MSE.

```python
# Practical 24
from sklearn.svm import SVR
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler

X, y = fetch_california_housing(return_X_y=True)
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3, random_state=42)

sc = StandardScaler().fit(Xtr)
Xtr_s, Xte_s = sc.transform(Xtr), sc.transform(Xte)

svr = SVR().fit(Xtr_s, ytr)
print("MAE:", round(mean_absolute_error(yte, svr.predict(Xte_s)),4))
```

##Practical 26. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
# Practical 26
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_curve, average_precision_score

X, y = load_breast_cancer(return_X_y=True)
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

clf = SVC(kernel='rbf', probability=True).fit(Xtr, ytr)
probs = clf.predict_proba(Xte)[:,1]
precision, recall, _ = precision_recall_curve(yte, probs)
ap = average_precision_score(yte, probs)

plt.plot(recall, precision, label=f'AP={ap:.3f}')
plt.xlabel('Recall'); plt.ylabel('Precision'); plt.title('Precision-Recall Curve')
plt.legend(); plt.show()
```
